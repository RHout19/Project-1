```python
#Exercise 3.1 | With interactive steps
genres = mv['genre'].unique()
range(1, 11)
x = ''
y = ''

#Find the genre
while x not in genres:
    x = input("Select Genre:").capitalize()
    if x not in genres:
        print(f"Please selection from {genres}")

#Find the rating
while True:
    y = input("Select Rating between 1-10")
    if y.isdigit() and int(y) in range(1, 11):
        y = int(y)
        break
    else:
        print("Please select a number between 1-10.")

mv_animation_7up = mv[(mv['genre'] == x) & (mv['rating'] > y)].shape[0]

if mv_animation_7up == 1:
    print(f"There is {mv_animation_7up} {x} movie with a rating above {y}.")
else:
    print(f"There are {mv_animation_7up} {x} movies with a rating above {y}.")
```

```python
# Exercise 9 | most & least frequent MPAA Rating in the dataset in terms of occurances

count = mv.value_counts('mpaa_rating')

most = count.index[0]
least = count.index[-1]

print(f'The most common rating is {most} and the least common rating is {least}.')
```

```python
# Exercise 10 | most and least expensive Genre (take an average of all Budget measures grouped by Genre - use groupBy() method)

expense_genre = mv.groupby('genre')['budget'].mean().nlargest(16)

most = expense_genre.index[0]
least = expense_genre.index[-1]

print(f'The most expensive genre is {most}, while the least cheapest genre is {least}.')
```

```python
# Exercise 11 | whic Genre is favored the most by the people?

most = mv.groupby('genre')['rating'].mean().nlargest(16).index[0]
least = mv.groupby('genre')['rating'].mean().nlargest(16).index[-1]

print(f'The most populaire genre is {most}, while the least populaire genre is {least}.')
```
