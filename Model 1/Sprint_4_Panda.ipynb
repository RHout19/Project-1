```python
#Exercise 3 | With interactive steps
genres = mv['genre'].unique()
range(1, 11)
x = ''
y = ''

#Find the genre
while x not in genres:
    x = input("Select Genre:").capitalize()
    if x not in genres:
        print(f"Please selection from {genres}")

#Find the rating
while True:
    y = input("Select Rating between 1-10")
    if y.isdigit() and int(y) in range(1, 11):
        y = int(y)
        break
    else:
        print("Please select a number between 1-10.")

mv_animation_7up = mv[(mv['genre'] == x) & (mv['rating'] > y)].shape[0]

if mv_animation_7up == 1:
    print(f"There is {mv_animation_7up} {x} movie with a rating above {y}.")
else:
    print(f"There are {mv_animation_7up} {x} movies with a rating above {y}.")
```
